Mengimpor library Python `matplotlib.pyplot` untuk membuat grafik

```python
import matplotlib.pyplot as plt
```

Mendeklarasikan variabel `nilai_mahasiswa` berupa list kosong

```python
nilai_mahasiswa = []
```

Melakukan perulangan sebanyak 10 kali untuk meminta pengguna memasukkan angka dan menyimpannya ke dalam list `nilai_mahasiswa`.

```python
print("====== Nilai Mahasiswa ======")
for i in range(10):
    angka = int(input(f"Masukkan nilai mahasiswa ke-{i+1}: "))
    nilai_mahasiswa.append(angka)
```

Melakukan perulangan untuk menghitung jumlah mahasiswa yang lulus (nilai ≥ 60) dan tidak lulus berdasarkan data pada list `nilai_mahasiswa`.

```python
jumlah_lulus = 0
jumlah_tidak_lulus = 0

for a in range(len(nilai_mahasiswa)):
    if nilai_mahasiswa[a] >= 60:
        jumlah_lulus += 1

for b in range(len(nilai_mahasiswa)):
    if nilai_mahasiswa[b] <= 60:
        jumlah_tidak_lulus += 1
```

Menampilkan data nilai mahasiswa, nilai tertinggi, nilai terendah, rata-rata, serta jumlah mahasiswa yang lulus. Kemudian membuat dua grafik batang untuk menampilkan perbandingan nilai tertinggi dan terendah, serta data kelulusan mahasiswa.

```python
print("Array:", nilai_mahasiswa)
print("Nilai maksimum:", max(nilai_mahasiswa))
print("Nilai minimum:", min(nilai_mahasiswa))
print("Rata-rata:", sum(nilai_mahasiswa)/len(nilai_mahasiswa))
print("jumlah lulus:", jumlah_lulus)

plt.bar(["terendah", "tertinggi"], [min(nilai_mahasiswa), max(nilai_mahasiswa)])
plt.title("Nilai Tertinggi/Terendah")
plt.xlabel("Indeks")
plt.ylabel("Nilai")
plt.show()

plt.bar(["lulus","tidak Lulus"], [jumlah_lulus,jumlah_tidak_lulus])
plt.title("Data Kelulusan")
plt.xlabel("Indeks")
plt.ylabel("Jumlah Lulus")
plt.show()
```
